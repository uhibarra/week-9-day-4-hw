```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head()
```

```python
# 1) How many neighborhood groups are available and which shows up the most?

air_bnb.groupby(by='neighbourhood_group').describe()

# 5 neighborhood groups are available
# Manhattan shows up the most
```

```python
# 2) Are private rooms the most popular in manhattan?

room_types = air_bnb.groupby(['room_type','neighbourhood_group']).describe()
display(room_types)

# Manhattan Room Type Counts
# Entire home/apt = 13199
# Private room = 7982
# Shared room = 480
# Private rooms are not the most popular in Manhattan
```

```python
# 3) Which hosts are the busiest and based on their reviews?

bnb_hosts = air_bnb.groupby(['host_name']).mean().sort_values('reviews_per_month', ascending=False)
bnb_hosts[['reviews_per_month']].head(20)

# Row NYC is the busiest based on their average reviews per month
```

```python
# 4) Which neighorhood group has the highest average price?

avg_price = air_bnb.groupby(['neighbourhood_group']).mean().sort_values('price', ascending=False)
avg_price[['price']]

# Manhattan has the highest average price
```

```python
# 5) Which neighbor hood group has the highest total price?

total_price = air_bnb.groupby(['neighbourhood_group']).sum().sort_values('price', ascending=False)
total_price[['price']]

# Manhattan has the highest total price
```

```python
# 6) Which top 5 hosts have the highest total price?

host_total_price = air_bnb.groupby(['host_name']).sum().sort_values('price', ascending=False)
host_total_price[['price']].head(5)

# Top 5 host with the highest price listed below
# Sonder(NYC), Blueground, Michael, David, Alex
```

```python
# 7) Who currently has no (zero) availability with a review count of 100 or more?

# Make 2 new dataframes: 1 for reviews and 1 for availability
reviews = air_bnb[['host_name', 'number_of_reviews']]
avail = air_bnb[['host_name', 'availability_365']]

# Filter out any reviews below 100, and any availability that is not 0
reviews_100 = reviews[reviews['number_of_reviews'] > 100]
avail_0 = avail[avail['availability_365'] == 0]

# Merge the two dataframes
rev_avail = reviews_100.merge(avail_0, how='inner', on='host_name')

# Drop duplicate host names
rev_avail_dropped = rev_avail.drop_duplicates(subset='host_name').reset_index(drop=True)
rev_avail_dropped

# 859 hosts have over 100 reviews and 0 availability
```

```python
# 8) What host has the highest total of prices and where are they located?

host_total_price = air_bnb.groupby(['host_name', 'neighbourhood_group']).sum().sort_values('price', ascending=False)
host_total_price[['price']].head(1)

# Sonder(NYC) has the highest total prices and they're located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?

location = air_bnb[['host_name', 'last_review', 'neighbourhood_group']]

queens = location[location['neighbourhood_group'] == 'Queens']

danielle = queens[queens['host_name'] == 'Danielle'].sort_values('last_review', ascending=False)
danielle

# Danielle from Queens last received a review in July 8, 2019
```
